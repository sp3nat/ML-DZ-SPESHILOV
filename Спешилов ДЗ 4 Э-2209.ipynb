```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# определяем функцию f(x) и её производную f'(x)
def f(x):
    return 3 * x**2

def df(x):
    return 6 * x
```

```python
# Реализация метода градиентного спуска для одной переменной
def gradient_descent_1d(x0, learning_rate, tol=1e-8, max_iter=100):
    x = x0
    history = []  # для сохранения значений x и f(x) на каждой итерации
    for i in range(max_iter):
        history.append((x, f(x)))
        x_new = x - learning_rate * df(x)
        if np.abs(x_new - x) < tol:
            break
        x = x_new
    return np.array(history)
```

```python
x0 = 5.0
learning_rate = 0.1
result_1d = gradient_descent_1d(x0, learning_rate)
```

```python
print("Градиентный спуск для функции одной переменной:")
for i, (xi, fxi) in enumerate(result_1d):
    print(f"Итерация {i}: x = {xi:.4f}, f(x) = {fxi:.4f}")
```

```python
#график сходимости
x_vals = np.linspace(-5, 5, 400)
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f(x_vals), label="f(x) = 3*x^2")
plt.scatter(result_1d[:, 0], result_1d[:, 1], color='red', label="Шаги спуска")
plt.plot(result_1d[:, 0], result_1d[:, 1], color='red', linestyle='--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Градиентный спуск (одномерная функция)")
plt.legend()
plt.grid(True)
plt.show()
```


***Двух переменных***

```python
# Определяем функцию f(x, y) и её градиент
def f2(x, y):
    return 3 * x**2 + 5 * y**2

def grad_f2(x, y):
    return np.array([6 * x, 10 * y])
```

```python
# Реализация метода градиентного спуска для двух переменных
def gradient_descent_2d(x0, y0, learning_rate, tol=1e-8, max_iter=100):
    x, y = x0, y0
    history = []  # для сохранения (x, y, f(x,y)) на каждой итерации
    for i in range(max_iter):
        history.append((x, y, f2(x, y)))
        grad = grad_f2(x, y)
        x_new = x - learning_rate * grad[0]
        y_new = y - learning_rate * grad[1]
        if np.linalg.norm(np.array([x_new - x, y_new - y])) < tol:
            break
        x, y = x_new, y_new
    return np.array(history)
```

```python
x0_2d, y0_2d = 5.0, 10.0
learning_rate_2d = 0.1
result_2d = gradient_descent_2d(x0_2d, y0_2d, learning_rate_2d)
print("\nГрадиентный спуск для функции двух переменных:")
for i, (xi, yi, fxy) in enumerate(result_2d):
    print(f"Итерация {i}: x = {xi:.4f}, y = {yi:.4f}, f(x,y) = {fxy:.4f}")

#построение 3D графика поверхности функции и траектории спуска
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

#создаем сетку для построения поверхности функции
X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = f2(X, Y)

ax.plot_surface(X, Y, Z, alpha=0.5, cmap='viridis')
ax.plot(result_2d[:, 0], result_2d[:, 1], result_2d[:, 2], marker='o', color='red', label="Путь спуска")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.set_title("Градиентный спуск (двумерная функция)")
plt.legend()
plt.show()
```

```python

from scipy.optimize import minimize_scalar

def f1(x):
    return 3 * x**2

def df1(x):
    return 6 * x

def steepest_descent_1d(x0, tol=1e-8, max_iter=100):
    """
    Реализует метод наискорейшего спуска для функции одной переменной с подбором оптимального шага (линейный поиск).

    Аргументы:
      x0       - начальное значение x
      tol      - критерий сходимости по изменению x
      max_iter - максимальное число итераций

    Возвращает:
      history  - массив, где каждая строка содержит [x, f(x)] для каждой итерации.
    """
    x = x0
    history = [(x, f1(x))]
    for i in range(max_iter):
        grad = df1(x)
        if np.abs(grad) < tol:
            break
        res = minimize_scalar(lambda lam: f1(x - lam * grad))
        lambda_opt = res.x
        x = x - lambda_opt * grad
        history.append((x, f1(x)))
    return np.array(history)

x0 = 5.0
history_1d = steepest_descent_1d(x0)

print("Метод наискорейшего спуска для функции одной переменной:")
for i, (xi, fxi) in enumerate(history_1d):
    print(f"Итерация {i}: x = {xi:.6f}, f(x) = {fxi:.6f}")

x_vals = np.linspace(-5, 5, 400)
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f1(x_vals), label="f(x) = 3*x^2")
plt.scatter(history_1d[:,0], history_1d[:,1], color='red', label="Шаги спуска")
plt.plot(history_1d[:,0], history_1d[:,1], color='red', linestyle='--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Наискорейший спуск (одномерная функция)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def f2(x, y):
    return 3 * x**2 + 5 * y**2

def grad_f2(x, y):
    return np.array([6 * x, 10 * y])

def steepest_descent_2d(x0, tol=1e-8, max_iter=100):
    """
    Реализует метод наискорейшего спуска для функции двух переменных с подбором оптимального шага (линейный поиск).

    Аргументы:
      x0       - начальное значение [x0, y0] (список или массив)
      tol      - критерий сходимости (норма градиента)
      max_iter - максимальное число итераций

    Возвращает:
      history  - массив, где каждая строка содержит [x, y, f(x,y)] для каждой итерации.
    """
    x = np.array(x0, dtype=float)
    history = [(x[0], x[1], f2(x[0], x[1]))]
    for i in range(max_iter):
        grad = grad_f2(x[0], x[1])
        if np.linalg.norm(grad) < tol:
            break
        def f_line(lam):
            new_x = x - lam * grad
            return f2(new_x[0], new_x[1])
        res = minimize_scalar(f_line)
        lambda_opt = res.x
        x = x - lambda_opt * grad
        history.append((x[0], x[1], f2(x[0], x[1])))
    return np.array(history)

x0_2d = [5.0, 10.0]
history_2d = steepest_descent_2d(x0_2d)

print("\nМетод наискорейшего спуска для функции двух переменных:")
for i, (xi, yi, fxy) in enumerate(history_2d):
    print(f"Итерация {i}: x = {xi:.6f}, y = {yi:.6f}, f(x,y) = {fxy:.6f}")

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = f2(X, Y)
ax.plot_surface(X, Y, Z, alpha=0.5, cmap='viridis')
ax.plot(history_2d[:, 0], history_2d[:, 1], history_2d[:, 2],
        marker='o', color='red', label="Путь спуска")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.set_title("Наискорейший спуск (двумерная функция)")
plt.legend()
plt.show()
```
